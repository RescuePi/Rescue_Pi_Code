```python
# Import the needed packages for model training and conversion
from fastbook import *
from fastai.vision.widgets import *
import cv2
import torch
import torch.onnx
import onnx
```

## Load the Dataset

```python
# Set all constants
dataset_path = '/home/srinivassriram/Desktop/Rescue_Pi_Code/Rescue_Pi_Code/Dataset2/'
```

```python
# Build a callable instance of the DataBlock API
model_data = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
# Instantiate the Datablock using the imported dataset
dls = model_data.dataloaders(dataset_path, bs=64)
```

```python
# Perform data augmentation on the dataset: First, build a callable instance of the aug_transforms API inside of DataBlock 
model_aug = model_data.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
```

```python
# Reinitialize the dataset using the data augmentation API called previously.
dls = model_aug.dataloaders(dataset_path)
# Show a random training batch of what our dataset looks like with the augmentation.
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

## Train the Model

```python
# Build our CNN Model: Using our defined Dataset through the DataBlock API, ResNet18 Architecture, and for metrics, validation error rate and accuracy.
learn = cnn_learner(dls, resnet34, metrics=[accuracy, error_rate])
```

```python
# Fine tune our model on a pretrained network.
learn.fit(5)
# Did not run for many epochs due to lack of time and not wanting the model to overfit.
```

## Visualize the Results

```python
# Create a confusion matrix to analyze the validation classifications while training.
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
# Display the top losses of the model
interp.plot_top_losses(5, nrows=5)
```

```python
# Display the classification report
interp.print_classification_report()
```

## Convert PyTorch Model to ONNX Model (for Boosted Inference Speeds)

```python
# Define the necessary constants for model conversion
image_sz = 128 # Used while training the model
# Dummy data and inputs needed for converting the model to ONNX
dummy_data = torch.randn(1, 3, image_sz, image_sz)
dummy_input = torch.autograd.Variable(dummy_data)
# Naming conventions for converted model
output_model_name = 'Final_Rescue_Model_Onnx.onnx'
input_name = 'image'
output_name = 'warn'
```

```python
# Convert the model using the constants defined above and torch's onnx export function
torch.onnx.export(model, dummy_input, output_model_name, input_names = [input_name], output_names = [output_name])
```

```python
# Check to make sure that model conversion went successfully and did not produce an invalid model.
onnx_model = onnx.load(output_model_name)
onnx.checker.check_model(onnx_model)
```
